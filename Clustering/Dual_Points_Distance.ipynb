```python
import pandas as pd
import geopandas as gpd
```

```python
data = gpd.read_file("C:/Users/sohra/Desktop/Dr_Baradaran/Bardaskan_Sabzevar/Temp/Tasadofat.shp",)
```

```python
data = data.sort_values(by=["Y_"],ascending=False,ignore_index=True)
data.head()
```

```python
# Assigning ID numbers to the Accidents which follows a North-South direction along the Sabzevar-Bardaskan Road
for i in range(0,data.index.values[-1]+1):
    if data.loc[i,("ID")] == -1:   # ID numbers equal to -1 are points that have been added to the dataset (one point at every 100 meters) to increase the accuracy of calculating the Distance between each pair of successive accidents 
        pass
    elif (i - sum(data.loc[0:i-1,("ID")] == -1)) == data.loc[i-1,("ID")]:
        data.loc[i,("ID")] = data.loc[i-1,("ID")] + 1   # Avoiding chained indexing
    else:
        data.loc[i,("ID")] = i - sum(data.loc[0:i-1,("ID")] == -1) + 1
```

```python
data["Distance"] = 0
```

```python
# Calculating the distance between each pair of successive points (both the accident and added points)
for i in range(1,len(data)):
    data.loc[i,("Distance")] = data.loc[i-1,("geometry")].distance(data.loc[i,("geometry")])
```

```python
# Calculate the distance between the two successive accidents
data["Dual_Distance"] = 0
j = 0

for i in range(0,len(data)):
    if (data.loc[i,("ID")] == -1) or (data.loc[i,("ID")] == 1):
        print("passed index: ",i)   # Printing logs to debug possible mistakes
        pass
    
    else:
        j += 1
        
        # Printing logs to debug possible mistakes
        print("ID: ",j)
        print(data.loc[data.loc[data["ID"] == j].index.values[0]+1:i,("Distance")])
        print("Distance = ",sum(data.loc[data.loc[data["ID"] == j].index.values[0]+1:i,("Distance")]))
        
        # Suming up the length of the segments which locate between the successive accidents
        data.loc[i,("Dual_Distance")] = sum(data.loc[data.loc[data["ID"] == j].index.values[0]+1:i,("Distance")])
        
```
